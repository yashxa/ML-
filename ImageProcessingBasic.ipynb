```python
# convert rgb image into gray scale
import cv2
import matplotlib.pyplot as plt

# Read the RGB image
img_rgb = cv2.imread('/content/drive/MyDrive/images/apple.jpg')

# Convert RGB image to grayscale
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)
# img_gray = cv2.imread('/content/drive/MyDrive/images/apple.jpg',0)
# Plot both images side by side
plt.figure(figsize=(10, 5))

# Original RGB image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img_rgb, cv2.COLOR_BGR2RGB))
plt.title('Original RGB Image')
plt.axis('off')

# Grayscale image
plt.subplot(1, 2, 2)
plt.imshow(img_gray, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Read the image
img = cv2.imread('/content/drive/MyDrive/images/apple.jpg')

# Split the channels
blue_channel, green_channel, red_channel = cv2.split(img)

# Display the extracted channel
plt.figure(figsize=(10, 5))

# Display the blue channel
plt.subplot(1, 3, 1)
plt.imshow(blue_channel, cmap='gray')
plt.title('Blue Channel')
plt.axis('off')

# Display the green channel
plt.subplot(1, 3, 2)
plt.imshow(green_channel, cmap='gray')
plt.title('Green Channel')
plt.axis('off')

# Display the red channel
plt.subplot(1, 3, 3)
plt.imshow(red_channel, cmap='gray')
plt.title('Red Channel')
plt.axis('off')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Read the image
img = cv2.imread('/content/drive/MyDrive/images/apple.jpg')

# Convert the image to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Calculate the histogram
hist = cv2.calcHist([gray_img], [0], None, [256], [0,256])

# Plot the histogram
plt.figure(figsize=(8, 6))
plt.plot(hist, color='black')
plt.title('Grayscale Image Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Read the image
img = cv2.imread('/content/drive/MyDrive/images/apple.jpg')
# img=cv2.resize(img,(50,50))
# Define different interpolation methods
interpolation_methods = [cv2.INTER_NEAREST, cv2.INTER_LINEAR, cv2.INTER_CUBIC, cv2.INTER_LANCZOS4]
method_names = ['Nearest Neighbor', 'Bilinear', 'Bicubic', 'Lanczos']

# Plot the original image
plt.figure(figsize=(15, 5))
plt.subplot(1, 5, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original')
plt.axis('off')

# Apply different interpolation methods and plot the results
for i, method in enumerate(interpolation_methods):
    resized_img = cv2.resize(img, (50, 50), interpolation=method)
    plt.subplot(1, 5, i+2)
    plt.imshow(cv2.cvtColor(resized_img, cv2.COLOR_BGR2RGB))
    plt.title(method_names[i])
    plt.axis('off')

plt.show()
```

```python
import os
import cv2
import matplotlib.pyplot as plt

def read_images_from_folder(folder_path, resize=None):
    images = []
    for filename in os.listdir(folder_path):
        if filename.endswith(('.png', '.jpg', '.jpeg')):
            # Read the image using OpenCV
            image_path = os.path.join(folder_path, filename)
            img = cv2.imread(image_path)

            if img is not None:
                # Resize the image if resize argument is provided
                if resize is not None:
                    img = cv2.resize(img, resize)
                images.append(img)
            else:
                print(f"Unable to read image: {filename}")

    return images

def plot_images(images):
    num_images = len(images)
    rows = int(num_images / 3) + (num_images % 3 > 0)
    cols = min(3, num_images)

    fig, axes = plt.subplots(rows, cols, figsize=(12, 8))

    for i, ax in enumerate(axes.flatten()):
        if i < num_images:
            ax.imshow(cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB))
            ax.axis('off')
        else:
            fig.delaxes(ax)

    plt.show()

def normalize_images(images):
    normalized_images = []
    for img in images:
        # Normalize the image
        normalized_img = cv2.normalize(img, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_8U)
        normalized_images.append(normalized_img)
    return normalized_images

# Replace 'your_folder_path' with the actual path to your image folder
folder_path = '/content/drive/MyDrive/images'
images = read_images_from_folder(folder_path, resize=(100, 100))

# Normalize all the images
normalized_images = normalize_images(images)

# Plot all the normalized images
plot_images(normalized_images)
```
